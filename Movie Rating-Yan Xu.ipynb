```python
# Part One: Data loading and Initial Accessment
# Import Data from Github and check the first 10 lines
import pandas as pd
url = 'https://raw.githubusercontent.com/sundeepblue/movie_rating_prediction/master/movie_metadata.csv'
rating = pd.read_csv(url,parse_dates=[0])

rating.head(10)
```

```python
# Quick summary of numerical data
rating.describe()
```

```python
# Check the data type of each feature
print(rating.dtypes)
```

```python
#Display several features that was hidden from the head view
rating.loc[0:9, ['facenumber_in_poster', 'plot_keywords','num_voted_users','num_user_for_reviews']] 
```

```python
#Eliminate columns that not neccessary for analysis
rating1 = rating # make a copy of the original dataset
rating1.drop(['gross', 'movie_imdb_link','plot_keywords','facenumber_in_poster','language'], axis=1, inplace=True)
rating1.head(10)
```

```python
# Summarize missing values for each feature
print(rating1.isnull().sum())
```

```python
# Change certain object data type to categorical data type

rating1[['color','genres','country','content_rating','title_year','director_name', 'actor_2_name', 'actor_1_name','actor_3_name']] = rating1[['color','genres','country','content_rating','title_year','director_name', 'actor_2_name', 'actor_1_name','actor_3_name']].astype('category')
```

```python
print(rating1.dtypes)
```

```python
# Part Two: Data Cleaning 
# Part A. Replacing missing categorical elements as a Missing category 
# Replacing missings in director_name with Missing category
rating1['director_name'] = rating1['director_name'].cat.add_categories('Missing')
rating1['director_name'].fillna('Missing', inplace = True) 
# Replacing missings in actor_2_name  with Missing category
rating1['actor_2_name'] = rating1['actor_2_name'].cat.add_categories('Missing')
rating1['actor_2_name'].fillna('Missing', inplace = True) 
# Replacing missings in actor_1_name  with Missing category
rating1['actor_1_name'] = rating1['actor_1_name'].cat.add_categories('Missing')
rating1['actor_1_name'].fillna('Missing', inplace = True)
# Replacing missings in actor_3_name  with Missing category
rating1['actor_3_name'] = rating1['actor_3_name'].cat.add_categories('Missing')
rating1['actor_3_name'].fillna('Missing', inplace = True) 
# Replacing missings in country  with Missing category
rating1['country'] = rating1['country'].cat.add_categories('Missing')
rating1['country'].fillna('Missing', inplace = True) 
# Replacing missings in content_rating  with Missing category
rating1['content_rating'] = rating1['content_rating'].cat.add_categories('Missing')
rating1['content_rating'].fillna('Missing', inplace = True) 
# Replacing missings in title_year  with Missing category
rating1['title_year'] = rating1['title_year'].cat.add_categories('Missing')
rating1['title_year'].fillna('Missing', inplace = True) 





```

```python
# check the number of missing values after filling Missing category to categorical features
print(rating1.isnull().sum())
```

```python
# Data Cleaning
# Part B. Impute missing data in float data type with mean value of each feature
rating1['num_critic_for_reviews'] = rating1['num_critic_for_reviews'].fillna(rating1['num_critic_for_reviews'].mean())
rating1['duration'] = rating1['duration'].fillna(rating1['duration'].mean())
rating1['director_facebook_likes'] = rating1['director_facebook_likes'].fillna(rating1['director_facebook_likes'].mean())
rating1['actor_3_facebook_likes'] = rating1['actor_3_facebook_likes'].fillna(rating1['actor_3_facebook_likes'].mean())
rating1['actor_1_facebook_likes'] = rating1['actor_1_facebook_likes'].fillna(rating1['actor_1_facebook_likes'].mean())
rating1['num_user_for_reviews'] = rating1['num_user_for_reviews'].fillna(rating1['num_user_for_reviews'].mean())
rating1['budget'] = rating1['budget'].fillna(rating1['budget'].mean())
rating1['actor_2_facebook_likes'] = rating1['actor_2_facebook_likes'].fillna(rating1['actor_2_facebook_likes'].mean())

#Impute missing data for aspect_ratio with mode which is 2.35
rating1['aspect_ratio'].fillna(2.35,inplace=True)

```

```python
# check the number of missing values after imputing numerical values
print(rating1.isnull().sum())
```

```python
# Part Four: Modeling
# Part A: Logistic Regression 
# Since the correlation value between actor_3_facebook_likes and imdb_score, actor_1_facebook_likes and imdb_score,
# cast_total_facebook_likes and imdb_score, budget and imdb_score, actor_2_facebook_likes and imdb_score, aspect_ratio and imdb_score
# are relatively low, those values will be elimated from the analysis. Also, most of the movie are in color, therefore, we can elimate 
# color feature.
rating1.drop(['actor_3_facebook_likes', 'actor_1_facebook_likes','cast_total_facebook_likes','budget','actor_2_facebook_likes','aspect_ratio','color'],axis=1, inplace=True)
rating1.head(10)
```

```python
rating1.shape
```

```python
# Select only numerical features and leave out the dependent feature - imdb_score
rating1_selected = rating1[['num_critic_for_reviews','duration','director_facebook_likes','num_voted_users','num_user_for_reviews','movie_facebook_likes']]
rating1_selected.head(10)
```

```python
# For categorical values, need to use one-hot encoding to convert it workable format
dummy = pd.get_dummies(rating1[['director_name', 'actor_2_name','genres','actor_1_name','movie_title','actor_3_name','country','content_rating','title_year']])
dummy.head()
```

```python
# Form the dataset for model
model_data = pd.concat([rating1_selected, dummy], axis =1)
model_data.shape
```

```python
# Split data into Test Dataset and Train Dataset
from sklearn.model_selection import train_test_split
X = model_data
y = rating1.imdb_score
# Split X and y into X_
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21)
```

```python
from sklearn import preprocessing
lab_enc = preprocessing.LabelEncoder()
y_train_encoded = lab_enc.fit_transform(y_train)
```

```python
y_test_encoded = lab_enc.fit_transform(y_test)
```

```python
# Logistic Regression 
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,  y_train_encoded)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_test, y_test_encoded) * 100, 2)
acc_log
```

```python
# Trying to select features and improve accuracy score
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
def select_features(X_train, y_train_encoded, X_test):
    fs = SelectKBest(score_func=chi2, k=500)
    fs.fit(X_train, y_train_encoded)
    X_train_fs = fs.transform(X_train)
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs

X_train_fs, X_test_fs = select_features(X_train, y_train_encoded, X_test)

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train_fs,  y_train_encoded)
Y_pred = logreg.predict(X_test_fs)
acc_log_chi = round(logreg.score(X_test_fs, y_test_encoded) * 100, 2)
acc_log_chi
```

```python
# Run SVC
from sklearn.svm import SVC, LinearSVC
svc = SVC()
svc.fit(X_train,  y_train_encoded)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_test, y_test_encoded) * 100, 2)
acc_svc
```

```python
# Linear SVC
linear_svc = LinearSVC()
linear_svc.fit(X_train,  y_train_encoded)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_test,  y_test_encoded) * 100, 2)
acc_linear_svc
```

```python
# # Stochastic Gradient Descent
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier()
sgd.fit(X_train,  y_train_encoded)
Y_pred = sgd.predict(X_test)
acc_sgd = round(sgd.score(X_test,  y_test_encoded) * 100, 2)
acc_sgd
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=1000)
random_forest.fit(X_train,  y_train_encoded)
Y_pred = random_forest.predict(X_test)
acc_random_forest = round(random_forest.score(X_test,  y_test_encoded) * 100, 2)
acc_random_forest
```

```python
# Hyperparameter Tuning
# Create a parameter grid 
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf , param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train,  y_train_encoded)
print(rf_random.best_params_)
```

```python
acc_random_forest_tune = round(rf_random.score(X_test,  y_test_encoded) * 100, 2)
acc_random_forest_tune
```

```python
#K Nearest Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 100)
knn.fit(X_train,  y_train_encoded)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_test,  y_test_encoded) * 100, 2)
acc_knn
```

```python
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
gaussian.fit(X_train,  y_train_encoded)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_test,  y_test_encoded) * 100, 2)
acc_gaussian
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train,  y_train_encoded)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_test,  y_test_encoded) * 100, 2)
acc_decision_tree
```
